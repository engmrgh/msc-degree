```python
import os
import glob
import warnings
import cv2
import scipy
import numpy as np
import pandas as pd
import imageio as iio
import plotly.express as px
import plotly.graph_objects as go
from sklearn.metrics import confusion_matrix

EPSILON = 0.001
pd.options.plotting.backend = "plotly"
```

# Question 2

## Part A

```python
CLASSES = [0, 1]
```

Loading train dataset.

```python
train = {class_:list() for class_ in CLASSES}
for class_ in train.keys():
    for filepath in glob.glob(f'./data/P2/dataset/a/Train/{class_}*'):
        train[class_].append(255-iio.imread(filepath))
```

Corresponding image for each of the classes is shown below.

```python
os.makedirs('images/q2/parta/', exist_ok=True)
model = dict()
for class_ in CLASSES:
    model[class_] = np.mean(train[class_],axis=0)
    fig = px.imshow(model[class_], 
                    color_continuous_scale='gray',
                    title=f'Mean image of class {class_}')
    fig.update_layout(title_x=0.5)
    fig.write_image(f'images/q2/parta/{class_}_prototype.png')
    fig.show()
```

## Part B

Loading test data

```python
test = {class_:list() for class_ in CLASSES}
for class_ in CLASSES:
    for filepath in glob.glob(f'./data/P2/dataset/a/Test/{class_}*'):
        test[class_].append(255-iio.imread(filepath))
```

Evaluating model based on Test dataset.

```python
error = 0
num_of_instances = 0
for true_class in CLASSES:
    for instance in test[true_class]:
        num_of_instances += 1
        min_distance = float('inf')
        predicted_class = None
        for class_ in CLASSES:
            distance = np.sum(np.abs(instance - model[class_]))
            if distance < min_distance:
                min_distance = distance
                predicted_class = class_
        if true_class != predicted_class:
            fig = px.imshow(instance, color_continuous_scale='gray',
                            title=f'True Label {true_class}, Predicted Lable {predicted_class}')
            fig.update_layout(title_x=0.5)
            fig.show()
            error += 1
print('Number of instances misclassified:', error)
print('Percent of instances misclassified:', (error/num_of_instances))
```
