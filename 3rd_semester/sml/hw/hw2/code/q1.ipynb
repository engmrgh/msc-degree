```python
import numpy as np
from scipy.stats import norm
from matplotlib import pyplot as plt
```

```python
p = lambda x: norm(3, 0.5).pdf(x) + norm(7, 0.5).pdf(x)

q1 = lambda x: norm(3, 0.5).pdf(x)
q2 = lambda x: norm(5, 1.5).pdf(x)
q3 = lambda x: norm(7, 0.5).pdf(x)
```

```python
def kl_distance(p, q):
    linespace = np.linspace(-20, 20, 10_000)

    p_prob = p(linespace)    
    q_prob = q(linespace)
    X = (p_prob > 0) & (q_prob > 0)

    p_prob = p_prob[X]
    q_prob = q_prob[X]
    return np.sum(p_prob * np.log(p_prob / q_prob))
```

```python
kl_p_q1 = kl_distance(p, q1)
kl_p_q2 = kl_distance(p, q2)
kl_p_q3 = kl_distance(p, q3)

print(f"D_kl(p||q1)= {kl_p_q1:.4f}")
print(f"D_kl(p||q2)= {kl_p_q2:.4f}")
print(f"D_kl(p||q3)= {kl_p_q3:.4f}")
```

```python
kl_q1_p = kl_distance(q1, p)
kl_q2_p = kl_distance(q2, p)
kl_q3_p = kl_distance(q3, p)

print(f"D_kl(q1||p)= {kl_q1_p:.4f}")
print(f"D_kl(q2||p)= {kl_q2_p:.4f}")
print(f"D_kl(q3||p)= {kl_q3_p:.4f}")
```

```python
x = np.linspace(-2.5, 12.5, 10_000)
plt.plot(x, p(x), label="p")
plt.plot(x, q1(x), label="q1", linestyle="--")
plt.legend()
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('Distribution of q1 with respect to p')
plt.show()
```
