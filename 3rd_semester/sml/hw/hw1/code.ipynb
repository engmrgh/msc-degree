## Problem 3

```python
die = [1, 2, 3, 4, 5, 6]
A = {2, 3, 6}
B = {1, 2, 3, 4}

n_A = 0
n_B = 0
n_AB = 0
n_tot = 0

diff = list()
p_A_seq = list()
p_B_seq = list()
p_AB_seq = list()
for _ in range(1000):
    side_id = np.random.choice(len(die))
    side = die[side_id]

    n_tot += 1
    if side in A:
        n_A += 1
    if side in B:
        n_B += 1
    if side in A and side in B:
        n_AB += 1

    p_A = n_A / n_tot
    p_A_seq.append(p_A)
    p_B = n_B / n_tot
    p_B_seq.append(p_B)
    p_AB = n_AB / n_tot
    p_AB_seq.append(p_AB)
    diff.append((p_AB - p_A * p_B) ** 2)

f, (a0, a1) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3, 1]}, figsize=(12, 8))

a0.plot(np.arange(1, n_tot + 1), p_A_seq, label='P(A)')
a0.plot(np.arange(1, n_tot + 1), p_B_seq, label='P(B)')
a0.plot(np.arange(1, n_tot + 1), p_AB_seq, label='P(AB)')
a1.plot(np.arange(1, n_tot + 1), diff, label='P(AB) - P(A) * P(B)')
plt.suptitle('Difference between P(A and B) and P(A) * P(B)')
plt.xlabel('toss step')
a0.legend()
a1.legend()
plt.show()
```

```python
import numpy as np

n_success = 0
n_tries = 1000

door_set = [1, 2, 3]
for i in range(n_tries):
    prize_door = np.random.choice(door_set, size=1, p=[1/3, 1/3, 1/3])
    sel_door = np.random.choice(door_set, size=1, p=[1/3, 1/3, 1/3])
    if sel_door == prize_door:
        n_success += 1

print("P(success without chaning door) = ", n_success / n_tries)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def ecdf(a):
    x, counts = np.unique(a, return_counts=True)
    cusum = np.cumsum(counts)
    return x, cusum / cusum[-1]


def plot_ecdf(a, label):
    x, y = ecdf(a)
    x = np.insert(x, 0, x[0])
    y = np.insert(y, 0, 0.)
    plt.plot(x, y, drawstyle='steps-post', label=label)
    plt.grid(True)

plot_ecdf(x1+x2, label='X1+X2')

x_vals = np.linspace(700, 900, num=1000)
p_vals = x1_plus_x2.cdf(x_vals)
plt.plot(x_vals, p_vals, color='r', label='Bio(2000, 0.4)')
plt.legend()
```

```python
import numpy as np

p = 0.3
sigma = 0.2
n_tries = 1000
n_samples = 100

diff_list = list()
for _ in range(n_tries):
    X = np.random.choice([1, 0], size=n_samples, replace=True, p=[p, 1-p])
    X_bar = np.mean(X)
    diff_list.append(np.abs(X_bar-p))

diff_list = np.array(diff_list)
print(f"P(|X̄-p|>{sigma}) = {np.sum(diff_list>sigma)/n_tries}")
print(f"Chebishev's upperbound: {1/(4*n_samples*sigma**2):.3f}")
print(f"Hoeffding's upperbound: {2*np.exp(-2*n_samples*sigma**2):.3f}")
```
